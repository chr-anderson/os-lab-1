````python
import pandas as pd
import matplotlib.pyplot as plt
````

Reading the data from text file.
* The header is formatted inconveniently, so we skip it with ```skiprows = 1, header = None```

````python
#fp = "C:/Users/Student/Documents/UO/PHYS628/OpticStudio1/os-lab-1/interference_intensity_profiles"
filename4 = "MultipleData.txt"
filename2 = "DoubleData.txt"
data_4_slit = pd.read_csv(filename4, encoding = 'utf-16', skiprows = 1, header = None, sep = '\t| ', engine = 'python')
data_2_slit = pd.read_csv(filename2, encoding = 'utf-16', skiprows = 1, header = None, sep = '\t| ', engine = 'python')
````

Selecting just the middle row of this data (all pixels at y = 201) for the intensity plot, and generating a list of x values.

````python
def center_intensity_df(data):
    middle_row = pd.DataFrame(data.iloc[200,1:])
    middle_row.columns = ['intensity']
    x = [pixel / 500 for pixel in range(1, 402)]
    middle_row.insert(0, 'x', x)
    return(middle_row)
````

````python
intensity_4_slit = center_intensity_df(data_4_slit)
intensity_2_slit = center_intensity_df(data_2_slit)
````

````python
print(intensity_4_slit.shape)
intensity_4_slit.iloc[:5,:]
````

````python
plt.figure(dpi = 300, figsize = [6, 4])                            # 300 dpi, 6 x 4 inches

plt.plot(intensity_2_slit['x'], intensity_2_slit['intensity'],
         color = 'blue',
         lw = 0.75,
         alpha = 0.6)                                              # Double slit, 0.75 pt linewidth

plt.plot(intensity_4_slit['x'], intensity_4_slit['intensity'],
         color = 'green',
         lw = 0.80)                                                # Quadruple slit, 0.8 pt linewidth

plt.title("Intensity on Detector Plane at y = 0")                  # Title
plt.xlabel('X $(mm)$')                                             # Labels
plt.ylabel('Intensity $\\left ( \\frac{W}{cm^2} \\right )$')


plt.legend(['Double slit', 'Quadruple slit'],                      # Legend
           loc = 'upper left', bbox_to_anchor = (-0.01, 1.015),    # Adjust position from upper left
           frameon = False)                                        # No outline



plt.tight_layout()                                                 # Make room for labels before saving
plt.savefig("IntensityPlot.png")
````

Alternative. This takes up more vertical space and has some legend shenanigans to iron out.

````python
plt.figure(dpi = 300, figsize = [6, 6]) # 300 dpi, 6 x 4 inches

double_plot = plt.subplot(2, 1, 1)
plt.plot(intensity_2_slit['x'], intensity_2_slit['intensity'],
         color = 'blue',
         lw = 0.75) # Double slit, 0.75 pt linewidth

plt.ylabel('Intensity $\\left ( \\frac{W}{cm^2} \\right )$')


quad_plot = plt.subplot(2, 1, 2, sharey = double_plot)
plt.plot(intensity_4_slit['x'], intensity_4_slit['intensity'],
         color = 'green',
         lw = 0.75) # Quadruple slit, 0.75 linewidth

plt.xlabel('X $(mm)$')
plt.ylabel('Intensity $\\left ( \\frac{W}{cm^2} \\right )$')


plt.legend(['Double slit', 'Quadruple slit'],                        # Legend
           loc = 'lower center', bbox_to_anchor = (0.5, -0.4),       # Adjust position from lower center
           fancybox = False, edgecolor = 'black')                    # Black outline with no rounded corners



plt.tight_layout()                                                   # Make room for labels before saving
````
